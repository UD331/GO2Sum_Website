```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_function = 'Function_test_summary_score_res.tab'
file_pathway = 'Pathway_test_summary_score_res.tab'
file_subunit = 'Subunit_test_summary_score_res.tab'
```

```python

func = pd.read_csv(file_function,sep='\t')
print(func.head(5))
pathway = pd.read_csv(file_pathway,sep='\t')
sub = pd.read_csv(file_subunit,sep='\t')
```

```python
print('Here is the density graph for total score')
sns.set(style="darkgrid")
# plotting both distibutions on the same figure
fig = sns.kdeplot(func['sum'], fill=True, color="r")
fig = sns.kdeplot(pathway['sum'], fill=True, color="b")
fig = sns.kdeplot(sub['sum'], fill=True, color="g")
fig.set(title='Distribution of combined scores', xlabel='scores', ylabel='density')
plt.legend(title='Sum', loc='upper left', labels=['Function', 'Pathway','Subunit'])
plt.show()
```

```python
print('Here is the density graph for model(BERT) score')
sns.set(style="darkgrid")
# plotting both distibutions on the same figure
fig = sns.kdeplot(func['model(BERT)'], fill=True, color="r")
fig = sns.kdeplot(pathway['model(BERT)'], fill=True, color="b")
fig = sns.kdeplot(sub['model(BERT)'], fill=True, color="g")
fig.set(title='Distribution of model(BERT) scores', xlabel='scores', ylabel='density')
plt.legend(title='model(BERT)', loc='upper left', labels=['Function', 'Pathway','Subunit'])
plt.show()
```
